```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import validation_curve

from sklearn.pipeline import make_pipeline

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
rng=np.random.RandomState(1)
(rng.rand(40,1)**2).ravel()
```

```python
# Let's explore np and generate some data for fitting and predicting the model
def generate_data(N,err=1.0,rseed=1):
    # Randomly sample the data
    rng=np.random.RandomState(rseed)
    X=rng.rand(N,1)**2
    y=10-1./(X.ravel()+0.1)
    
    if err>0:
        y+=err*rng.randn(N)
        
    return X,y
```

```python
X,y=generate_data(40)
```

```python
# Let's plot this data to see the distribution
plt.scatter(X.ravel(),y,color='black')
plt.show()
```

```python
# Let's define a function that computes Polynomial Regression
def PolynomialRegression(degree=2,**kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                        LinearRegression(**kwargs))
```

```python
# Let's Visualise different degrees of the model that best fits the data

plt.scatter(X.ravel(),y,color='black')
X_test=np.linspace(-0.1,1.1,500)[:,np.newaxis]


for degree in [1,3,5]:
    y_test=PolynomialRegression(degree).fit(X,y).predict(X_test)
    plt.plot(X_test,y_test,label='degree={}'.format(degree))
    
plt.legend(loc='best')
plt.show()
```
