```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmod(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.arange(-7,7,0.1)
```

```python
phi_x=sigmod(x)
```

```python
plt.plot(x,phi_x)
plt.axvline(0.0,color='k')
plt.yticks([0.0,0.5,1.0])
plt.xlabel('Values of x')
plt.ylabel('sigmoid values')
ax=plt.gca()
ax.yaxis.grid()
plt.show()
```

## Plotting the cost function curve 
<img src=imgs/cost_fn_1.png>

```python
x_values=np.arange(-10,10,0.2)
phi_z=sigmod(x_values)
```

```python
def cost_fn_0(z):
    return -np.log(1-sigmod(z))

def cost_fn_1(z):
    return -np.log(sigmod(z))
```

```python
cost_0=[cost_fn_0(i) for i in x_values]

cost_1=[cost_fn_1(i) for i in x_values]
```

```python
plt.plot(phi_z,cost_0,linestyle='--',label='J(w) if y=0')
plt.plot(phi_z,cost_1,label='J(w) if y=1')
plt.xlabel('$\phi$(x)')
plt.ylabel('J(w)')
plt.xlim(0,1)
plt.ylim(0.0,5.0)
plt.legend(loc='best')
plt.show()
```

Hence, we can see that the cost f'n value is almost 0 for values we correctly predict as label 1 and same for label 0, where the value of cost f'n converges to 0 at  '0'
