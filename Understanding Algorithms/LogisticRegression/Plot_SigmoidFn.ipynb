```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmod(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.arange(-7,7,0.1)
```

```python
phi_x=sigmod(x)
```

```python
plt.plot(x,phi_x)
plt.axvline(0.0,color='k')
plt.yticks([0.0,0.5,1.0])
plt.xlabel('Values of x')
plt.ylabel('sigmoid values')
ax=plt.gca()
ax.yaxis.grid()
plt.show()
```
