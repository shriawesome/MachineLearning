```python
import numpy as np

from sklearn.datasets import make_blobs 

from KMeans import KMeans
```

## Generate the data using make_blobs

```python
# Creating a dataset
np.random.seed(42)
X ,y = make_blobs(centers=3, n_samples=500, n_features=2, shuffle=True, random_state=40)
X.shape
```

```python
np.unique(y)
```

## Visualise the training using custom K

```python
k = KMeans(k=3, plot_steps=True)
k.fit(X)
```

```python
# Make predictions
y_pred = k.predict(X)
```
