```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
%matplotlib inline

from sklearn import datasets

from sklearn.model_selection import train_test_split
from sklearn.manifold import Isomap
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
# Load the digits dataset
digits=datasets.load_digits()
digits.images.shape          # So we can see there are 1797 samples each in (8 X 8) grids
```

```python
# Let's plot first 100 of these images and understand the data
fig,axes=plt.subplots(10,10,figsize=(8,8),
                     subplot_kw={'xticks':[],'yticks':[]},
                     gridspec_kw=dict(hspace=0.1,wspace=0.1))

for i,ax in enumerate(axes.flat):
    ax.imshow(digits.images[i],cmap='binary')
    ax.text(0.05,0.05,str(digits.target[i]),transform=ax.transAxes,color='green')
```

```python
# Now to train the data we need to convert this in the from of [n_samples,n_features] format for 
# feature matrix and [n_samples,] for target matrix
X_digits=digits.data
y=digits.target
X_digits.shape,y.shape
```

# Dimensionality reduction

It it difficult to visualise the data in 64-dimension high space, so we'll reduce the dimensoin to 2D using an unsupervised method. For this task we'll use mainfold learning algorithm called `Isomap`

```python
iso=Isomap(n_components=2)
iso.fit(X_digits)
projected_data=iso.transform(X_digits)
projected_data.shape
```

```python
# Now let's plot this projected data using matplotlib
# plt.cm.cmap_d.keys(): can be used to check different colors option available

plt.scatter(projected_data[:,0],projected_data[:,1],c=digits.target,alpha=0.5,
           cmap=plt.cm.get_cmap('nipy_spectral',10))
plt.colorbar(label='digit label',ticks=range(10))
plt.clim(-0.5,9.5)
```

```python
# From the above plot we can see that the digits are quite seperable and can be seperated using 
# even a simple classification alogrithm. Hence, we'll use GaussianNB. But prior to that let's split the
# Data into training and testing

x_train,x_test,y_train,y_test=train_test_split(X_digits,y,test_size=0.3,random_state=1)

# Let's create the object of the class and fit it in the same line
clf=GaussianNB().fit(x_train,y_train)
```

```python
# let's make the predictions on the test dataset
y_pred=clf.predict(x_test)
```

```python
# let's now evaluate the performance of the model
accuracy_score(y_test,y_pred)
```

Hence, we can see that even a simple GaussianNB gives us the accuacy of 83%, but just using accuracy score is not a good measure to evaluate model performance. Therefore, we'll use confusion matrix as well to see how actually the data is classified.

```python
# indices of all the misclassified data
np.where(y_pred!=y_test)
```

```python
mat=confusion_matrix(y_test,y_pred)

sns.heatmap(mat,annot=True,square=True,cbar=False)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

Using the above heatmap we can check for the misclassified data and which classes are 
misclassified often.
