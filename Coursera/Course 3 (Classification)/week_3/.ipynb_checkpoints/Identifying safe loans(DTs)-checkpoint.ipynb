```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
loans=pd.read_csv('data_set/lending-club-data.csv')
loans.head()
```

```python
loans.columns
```

```python
loans['safe_loans']=loans['bad_loans'].apply(lambda x:+1 if x==0 else -1)
```

```python
loans=loans.drop(columns='bad_loans')
```

```python
loans.drop(columns=['bad_loans'],axis=1)
```

```python
loans.groupby('safe_loans').size()/loans.shape[0]*100
```

```python
loans.shape[0]
```

## Features to be used

```python
features = ['grade',                     # grade of the loan
            'sub_grade',                 # sub-grade of the loan
            'short_emp',                 # one year or less of employment
            'emp_length_num',            # number of years of employment
            'home_ownership',            # home_ownership status: own, mortgage or rent
            'dti',                       # debt to income ratio
            'purpose',                   # the purpose of the loan
            'term',                      # the term of the loan
            'last_delinq_none',          # has borrower had a delinquincy
            'last_major_derog_none',     # has borrower had 90 day or worse rating
            'revol_util',                # percent of available credit being used
            'total_rec_late_fee',        # total late fees received to day
           ]

target = 'safe_loans'   
```

```python
loans=loans[features+[target]]
```

```python
loans.columns
```

## Importing the training and the validation set

```python
train_idx=pd.read_json('data_set/module-5-assignment-1-train-idx.json')[0]
```

```python
train_data=loans.iloc[train_idx]
```

```python
validation_idx=pd.read_json('data_set/module-5-assignment-1-validation-idx.json')[0]
```

```python
validation_data=loans.iloc[validation_idx]
```

## Building a Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree_model=DecisionTreeClassifier(max_depth=6)
```

```python
x_train=np.array(train_data[features])
```

```python
y_train=np.array(train_data[target])
```

```python
decision_tree_model.fit(x_train,y_train)
```

```python
train_data.dtypes
```

```python
train_data['term_float']=train_data['term'].apply(lambda x:+1 if str(x)=='60 months' else 0)
```
